```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2, 4, 5])
```

```python
y = np.array([1.2, 2.8, 5.3])
```

```python
%matplotlib inline

plt.figure(figsize=(12, 7))
plt.title('x-y relationship')
plt.scatter(x,y)
plt.show()
```

```python
# model parameters
a0, a1 = 0, 0

# list containing cost in each epoch
lst_epoch_cost = list()

# learning rate (hyperparameter)
lr = 0.001

# shoudl we go for another epoch?
repeat =True

# loop index
j = 0

while repeat:
    
    # cost in an epoch
    epoch_cost = 0
    
    # partial derivatives w.r.t a0 and a1
    da0, da1 = 0, 0
    
    # decide whether another epoch is required
    if j > 0:
         if lst_epoch_cost[j-1] - lst_epoch_cost[j] < 0.00001:
                break
         else:
            # compute the gradient w.r.t model parameters
            for i in range (len(x)):
                da0 += -2*(y[i] - (a0 +a1*x[i]))
                da1 += -2*x[i]*(y[i] - (a0 +a1*x[i]))
    
            da0 /= len(x)
            da1 /= len(x)
    
    # update model paramters
    a0 -= da0*lr
    a1 -= da1*lr
    
    ### beginning of epoch
    
    # sum of squares in the current epoch
    for i in range(len(x)):
        
        epoch_cost += (y[i] - (a0 + a1*x[i]))**2
    
    epoch_cost /= len(x)
    
    # record the epoch cost
    lst_epoch_cost.append(epoch_cost)
    
    
    
    
                
    # increment loop index
    j += 1
    
    
        
```

```python
%matplotlib inline

plt.figure(figsize=(12, 7))
plt.title('gradient descent')
plt.scatter(np.arange(j)+1, lst_epoch_cost[0:j])
plt.show()
```

```python
y_pred = a0 + a1*x
```

```python
plt.figure(figsize=(7, 5))
plt.title('best fit line')
plt.scatter(x, y)
plt.plot(sorted(x), sorted(y_pred), color='red')
plt.show()
```

```python
import time
%matplotlib notebook

fig = plt.figure()
plt.xlim(-10, 10)
plt.ylim(-20, 30)
ax = fig.add_subplot(111)
plt.ion()


fig.show()
fig.canvas.draw()


# model parameters
a0, a1 = 0, 0

# list containing cost in each epoch
lst_epoch_cost = list()

# learning rate (hyperparameter)
lr = 0.001

# should we go for another epoch?
repeat = True

# loop index
j = 0

while repeat:
    
    # cost in this epoch
    epoch_cost = 0
    
    # partial derivatives w.r.t a0 and a1 in this epoch
    da0, da1 = 0, 0
    
    
    ### beginning of epoch

    # compute cost in the current epoch
    for i in range(len(x)):

        epoch_cost += (y[i] - (a0 + a1*x[i]))**2

    # record the epoch cost
    lst_epoch_cost.append(epoch_cost)
    
    # compute gradient w.r.t model parameters
    for i in range(len(x)):

        da0 += -2*(y[i] - (a0 + a1*x[i]))
        da1 += -2*x[i]*(y[i] - (a0 + a1*x[i]))

    # update model parameters
    a0 -= da0*lr
    a1 -= da1*lr

    if j > 0:
        if lst_epoch_cost[j-1] - lst_epoch_cost[j] < 0.00001:
            repeat = False
    
    j += 1
    
    ### end of epoch
    
    # predict values with x as input with the learned parameters
    y_pred = a0 + a1*x
    
    
    ax.clear()
    plt.xlim(0, 10)
    plt.ylim(0, 10)
    plt.title('X vs Y - model learning process - finding the best fit line')
    ax.plot(sorted(x), sorted(y_pred), color='red')
    ax.scatter(x, y)
    fig.canvas.draw()
    
    #time.sleep(1)
    
```
